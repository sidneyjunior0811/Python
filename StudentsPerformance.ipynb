```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/StudentsPerformance.csv')
```

```python
df = df.rename(columns={'race/ethnicity': 'ethnicity', 'parental level of education' : 'education', 'test preparation course' : 'course', 'math score' : 'math', 'reading score' : 'reading', 'writing score' : 'writing'})
```

```python
df.isnull().sum()
```

```python
# Separação por gênero 
mulher = df.loc[df.gender == 'female']
homem = df.loc[df.gender == 'male']
```

```python
homem.describe()
```

### **Separando dados utilizando a média**

```python
# Separando dados acima da média 68
mulher_media_maior = mulher.loc[(mulher['math']>=68) & (mulher['reading']>=68) & (mulher['writing']>=68)]
homem_media_maior = homem.loc[(homem['math']>= 68) & (homem['reading'] >= 68) & (homem['writing'] >= 68)]
```

```python
# Separando dados abaixo da média 68
mulher_media_abaixo = mulher.loc[(mulher['math'] < 68) & (mulher['reading'] < 68) & (mulher['writing'] < 68)]
homem_media_abaixo = homem.loc[(homem['math']< 68) & (homem['reading'] < 68) & (homem['writing'] < 68)]
```
