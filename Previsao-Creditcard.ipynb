```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
df = df[['Class','Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
       'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']]
```

### Separando dados para treino e teste

```python
x = df.drop('Class', axis = 1)
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
SEED = 42

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=SEED)
```

```python
print('x_train:',x_train.shape)
print('x_test:',x_test.shape)
print('y_train:',y_train.shape)
print('y_test:',y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from imblearn.metrics import specificity_score
```

```python
lr = LogisticRegression(max_iter=1000, random_state=SEED)
lr.fit(x_train, y_train)
previsao = lr.predict(x_test)

print('Acurácia: ', metrics.accuracy_score(y_test, previsao))
print('Previsão: ', metrics.precision_score(y_test, previsao))
print('Recal: ', metrics.recall_score(y_test, previsao))
print('F1: ', metrics.f1_score(y_test, previsao))
print('specificity:', specificity_score(y_test, previsao)) 
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.datasets import make_classification
from sklearn.metrics import ConfusionMatrixDisplay

%matplotlib inline

cm = confusion_matrix(y_test, previsao)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
y_pred_proba = lr.predict_proba(x_test)[::,1]
fpr, tpr,_ = metrics.roc_curve(y_test, y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)

plt.rcParams['figure.figsize'] = (12,8)
plt.plot(fpr, tpr, label = "LR, auc = "+str(auc))
plt.plot([0,1],[0,1],color='red', lw = 2, linestyle = '--')
plt.legend(loc = 4)
```

#### Metodo oversampling

```python
from imblearn.over_sampling import SMOTE
from imblearn.metrics import specificity_score
```

```python
SEED = 42
smote = SMOTE(random_state=SEED)

x = df.drop('Class', axis = 1)
y = df['Class']
```

```python
print(x.shape)
print(y.shape)
```

```python
x_resampled, y_resampled = smote.fit_resample(x,y)
```

```python
df_balanced = pd.concat([y_resampled, x_resampled], axis = 1)
df_balanced.head()
```

```python
df_balanced['Class'].value_counts()
```

```python
x = df_balanced.iloc[:,1:].values
y = df_balanced.iloc[:,0].values
```

```python
SEED = 42
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.25, random_state=SEED)
```

```python
lr = LogisticRegression(max_iter=1000, random_state=SEED)
lr.fit(x_train, y_train)
precisao = lr.predict(x_test)

print('acurácia:', metrics.accuracy_score(y_test, precisao))
print('Previsão:', metrics.precision_score(y_test, precisao))
print('Recall:', metrics.recall_score(y_test, precisao))
print('F1:', metrics.f1_score(y_test, precisao))
print('specificity:', specificity_score(y_test, precisao))
```

```python
cm = confusion_matrix(y_test, precisao)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
y_pred_proba = lr.predict_proba(x_test)[::,1]
fpr, tpr,_ = metrics.roc_curve(y_test, y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)

plt.rcParams['figure.figsize'] = (12,8)
plt.plot(fpr, tpr, label = "LR, auc = "+str(auc))
plt.plot([0,1],[0,1],color='red', lw = 2, linestyle = '--')
plt.legend(loc = 4)
```
