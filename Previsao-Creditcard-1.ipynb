```python
# Verificando dados de nÃ£o fraude 
df_nao_fraude = df_credito.Amount[df_credito.Class == 0]
df_nao_fraude.describe().round(2)
```

```python
# Verificando dados de fraude
df_fraude = df_credito.Amount[df_credito.Class !=0]
df_fraude.describe().round(2)
```

```python
df_fraude = df_credito[df_credito.Class == 1]
df_fraude
```

```python
df_nao_fraude = df_credito[df_credito.Class == 0]
display(df_nao_fraude)
```

```python
# Novo dataset com nÃºmero de nÃ£o fraude igual a de fraude
df_nao_fraude = df_nao_fraude.sample(n= 492)
display(df_nao_fraude)
```

```python
# Concatenando as linhas do Dataframe df_fraude e df_nao_fraude
df = pd.concat([df_fraude, df_nao_fraude], axis=0)
display(df)
```

```python
# Ajustando Index
df.reset_index(inplace= True)
df
```

```python
# Criando Variaveis com as 05(cinco) primeiras e ultimas linhas para validaÃ§Ã£o
df_Vali_Fraude = df.head(5)
df_Vali_nao_Fraude = df.tail(5)
```

```python
# Retirando os 05(cinco) primeiros e ultimas linhas
df = df.iloc[5:]
df = df[:-5]
df
```

```python
# Concatenando os dataframe de validaÃ§Ã£o
df_vali_total = pd.concat([df_Vali_Fraude, df_Vali_nao_Fraude])
df_vali_total.reset_index(inplace=True)

df_vali_total_real = df_vali_total.Class
df_vali_total = df_vali_total.drop(['level_0', 'index', 'Time', 'Class'], axis = 1)
df_vali_total
```

```python
# Verificando os dataframe fraude e nÃ£o fraude
df.Class.value_counts()
```

```python
# Separando labels e features
x = df.drop(['index', 'Time', 'Class'], axis = 1)
y = df['Class']
```

```python
# Separando dados para teste
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size= 0.2, random_state= 66, stratify= y)
```

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(x_train, y_train)
pred = lr.predict(x_test)
acuracia = accuracy_score(y_test, pred)

f'AcurÃ¡cia: {acuracia * 100:.2f}'
```

```python
# ValidaÃ§Ã£o
pred = lr.predict(df_vali_total)
df = pd.DataFrame({'Real': df_vali_total_real, 'PrevisÃ£o': pred})
print(df)
```
